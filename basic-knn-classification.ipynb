# K Nearest Neighbors 

In this notebook I use the KNN Alogrithm to classify anonymized data.

### Importing Libraries and Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../input/datasetknn/DataSetKNN.csv')
```

```python
df.head()
```

### Looking at the data

Since it's anonomized we can't really tell much from looking at it

```python
sns.pairplot(df,hue='TARGET CLASS',palette='coolwarm')
```

### Standardizing the variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
scaled_features = scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

### Splitting the Data in train and test data using train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['TARGET CLASS'],
                                                    test_size=0.30)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

### Predictions

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

### Using the elbow method to determine the k-value

```python
error_rate = []

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Wert')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

### Training the model again with k = 24

```python
knn = KNeighborsClassifier(n_neighbors=24)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('Mit K=24')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
